# 2. Ventas por categoría

```python
def generador_diccionario_para_listas(lista_categorias_tienda1: list = []):
  lista_categorias_unicos_tienda1 = []
  for categoria in lista_categorias_tienda1:
    if categoria not in lista_categorias_unicos_tienda1:
      lista_categorias_unicos_tienda1.append(categoria)
  #print(lista_categorias_unicos_tienda1)
  diccionario_categorias_tienda1 = {categoria_unica:lista_categorias_tienda1.count(categoria_unica) for categoria_unica in lista_categorias_unicos_tienda1}
  #print(diccionario_categorias_tienda1)
  return diccionario_categorias_tienda1

lista_categorias_tienda1 = tienda['Categoría del Producto'].tolist()
print(lista_categorias_tienda1)

diccionario_categorias_tienda1 = generador_diccionario_para_listas(lista_categorias_tienda1)
print(diccionario_categorias_tienda1)


```

```python
from matplotlib import pyplot as plt
import numpy as np

lista_productos_tienda1 = tienda['Producto'].tolist()
print(lista_productos_tienda1)

diccionario_productos_tienda1 = generador_diccionario_para_listas(lista_productos_tienda1)
print(diccionario_productos_tienda1)

def procesador_diccionarios(diccionario: dict = {}):
  tupla_productos_tienda = [(key,valor) for key,valor in diccionario.items()]
  print(tupla_productos_tienda)

  tupla_ordenada_tienda = sorted(tupla_productos_tienda, key = lambda x : x[1], reverse=True)
  return tupla_ordenada_tienda

tupla_ordenada_tienda1 = procesador_diccionarios(diccionario_productos_tienda1)

productos_mas_vendidos_tienda1 = tupla_ordenada_tienda1[:4]
print(productos_mas_vendidos_tienda1)

print("PRODUCTOS MAS VENDIDOS")
plt.bar([tupla[0] for tupla in productos_mas_vendidos_tienda1], [tupla[1] for tupla in productos_mas_vendidos_tienda1])
plt.xticks(rotation=45)
plt.show()

productos_menos_vendidos_tienda1 = tupla_ordenada_tienda1[-4:]
print(productos_menos_vendidos_tienda1)

print("PRODUCTOS MENOS VENDIDOS")

print("PRODUCTOS MENOS VENDIDOS")
plt.bar([tupla[0] for tupla in productos_menos_vendidos_tienda1], [tupla[1] for tupla in productos_menos_vendidos_tienda1])
plt.xticks(rotation=45)
plt.show()

#
```
